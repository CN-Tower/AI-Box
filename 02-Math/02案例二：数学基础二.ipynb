## 转移矩阵

```python
p = [
    [0.75, 0.125, 0.125],
    [0.5, 0.25, 0.25],
    [0.25, 0.5, 0.25]
]
```

```python
a = [0.5, 0.3, 0.2]
a = [0.1, 0.6, 0.3]
a = [0.0001, 0.9, 0.0999]
for i in range(300):
    t = []
    for j in range(3):
        tmp = p[0][j] * a[0] + p[1][j] * a[1] + p[2][j] * a[2]
        t.append(tmp)
    a = t
    t = list(map(lambda x: round(x, 4), t))
    print ("i=",i+1,"; a=",t)
```
